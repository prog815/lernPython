```
import matplotlib.pyplot as plt
import numpy as np
```

```
def rect_plot(n=5,a=np.array([0,0]),b=np.array([1,0]),c=np.array([1,1]),d=np.array([0,1]),alpha=0.25):
    """ рисуем вложенные квадраты """
    if n <= 0:
        return
    
    def plot_line(a,b):
        plt.plot([a[0],b[0]],[a[1],b[1]],c='red')

    plot_line(a,b)
    plot_line(b,c)
    plot_line(c,d)
    plot_line(d,a)

    def new_point(a,b):
        return a + alpha*(b-a)

    a1 = new_point(a,b)
    b1 = new_point(b,c)
    c1 = new_point(c,d)
    d1 = new_point(d,a)

    rect_plot(n-1,a1,b1,c1,d1,alpha)

plt.figure(figsize=(12,12))
rect_plot(n=50,alpha=0.9)
```
